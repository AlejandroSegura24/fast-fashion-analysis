```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importación del archivo csv

```python
file_path = 'https://raw.githubusercontent.com/AlejandroSegura24/fast-fashion-analysis/refs/heads/main/Data/true_cost_fast_fashion.csv'
data = pd.read_csv(file_path, encoding='windows-1252')
```

## Consultas basicas

### Top 5 paises con mayor producción de ropa

```python
country_more_production_clothes = data.groupby('Country')['Monthly_Production_Tonnes'].sum().sort_values(ascending=False).head(5)
country_more_production_clothes
```

### ¿Qué país genera más residuos textiles?

```python
country_res_total = data.groupby('Country')['Landfill_Waste_Tonnes'].sum().sort_values(ascending=False).head()
country_res_total
```

### Promedio de emisiones de CO₂ por país.

```python
avg_emision_co2 = data.groupby('Country')['Carbon_Emissions_tCO2e'].mean()
avg_emision_co2
```

```python
plt.figure(figsize=(12,6))
plt.plot(avg_emision_co2.index, avg_emision_co2.values, marker='o', linestyle='--',linewidth=2 , color='orange')
plt.grid()

plt.title('Promedio de CO2 por pais')
plt.xlabel('Pais')
plt.ylabel('Emision CO2')
```

## Analisis comparativo

### Comparar países desarrollados vs países en desarrollo (si hay categorización).

```python
#creando categorización de los paises desarrollados con los que estan en desarrollo
desarrollados = ['USA', 'UK', 'Germany', 'France', 'Japan', 'Canada']

data['Development'] = data['Country'].apply(lambda x: 'Desarrollado' if x in desarrollados else 'En desarrollo')
```

```python
data.info()
```

### Dependiendo el grupo de desarrollo, ¿Cuál grupo de paises produce mas ropa?

```python
country_development = data[data['Development']== 'Desarrollado']
country_not_development = data[data['Development']== 'En desarrollo']

prod_mensual_dev = country_development.groupby('Country')['Monthly_Production_Tonnes'].mean()
prod_mensual_no_dev = country_not_development.groupby('Country')['Monthly_Production_Tonnes'].mean()

fig, axs = plt.subplots(1, 2, figsize=(12, 4))

axs[0].bar(prod_mensual_dev.index, prod_mensual_dev.values, color='lightgreen')
axs[0].set_title('Promedio mensual por país desarrollado')
axs[0].tick_params(axis='x', rotation=45)

axs[1].bar(prod_mensual_no_dev.index, prod_mensual_no_dev.values, color='lightblue')
axs[1].set_title('Promedio mensual por país en desarrollo')
axs[1].tick_params(axis='x', rotation=45)
```

```python
development_avg = data.groupby('Development')['Monthly_Production_Tonnes'].mean()

plt.figure(figsize=(8,6))
labels = ['En desarrollo', 'Desarrollados']
colors = ['lightblue', 'lightgreen']

plt.pie(development_avg, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Promedio mensual de producción de ropa por Desarrollo')
plt.show()
```

### ¿Qué país tiene los hábitos de consumo más altos? (Segun su desarollo)

```python
dev_grouped = data.groupby('Development')['Avg_Spend_Per_Customer_USD'].mean()

dev_grouped_total = dev_grouped.sum()

print("Gasto por pais:\n", dev_grouped.to_string(), "\n")
print("Gasto total:", dev_grouped_total)
```

```python
plt.figure(figsize=(8,6))
labels = ['En desarrollo', 'Desarrollados']
colors = ['lightblue', 'lightgreen']

plt.pie(dev_grouped, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Promedio mensual de producción de ropa por tipo de país')
plt.show()
```

```python
country_not_development = data[data['Development']== 'En desarrollo']
not_dev_grouped = country_not_development.groupby('Country')['Avg_Spend_Per_Customer_USD'].mean()

not_dev_grouped_total = not_dev_grouped.sum()

print("Gasto por pais:\n", not_dev_grouped.to_string(), "\n")
print("Gasto total:", not_dev_grouped_total)
```

```python
development_or_not_development = data.groupby('Development')['Avg_Spend_Per_Customer_USD'].sum().sort_values(ascending=False)
print(development_or_not_development)

plt.subplots(figsize=(8,6))
labels = ['En desarrollo', 'Desarrollados']
colors = ['lightblue', 'lightgreen']

plt.pie(development_or_not_development, labels=labels, colors=colors, autopct='%1.1f%%',startangle=140)

plt.title('Compra de ropa segun el desarrollo de los paises')
plt.show()
```

## Relación entre consumo de ropa y emisiones de CO₂.

### Marca que produce mas CO2 y ropa al año

```python
Brand_produced_co2 = (
    data.groupby(['Year','Brand'])[['Monthly_Production_Tonnes', 'Carbon_Emissions_tCO2e']]
    .sum()
    .sort_values(by= 'Carbon_Emissions_tCO2e', ascending=False)
    .head(10)
    .reset_index()
)
Brand_produced_co2
```

```python
fig,axs = plt.subplots(1,2, figsize=(10, 4))

axs[0].barh(
    Brand_produced_co2['Brand'] + ' (' + Brand_produced_co2['Year'].astype(str) + ')',  # Concatenas marca + año como etiquetas
    Brand_produced_co2['Monthly_Production_Tonnes'],
    color='skyblue'
)
axs[0].set_title('Producción de ropa por marca-año')

axs[1].barh(
    Brand_produced_co2['Brand'] + ' (' + Brand_produced_co2['Year'].astype(str) + ')',
    Brand_produced_co2['Carbon_Emissions_tCO2e'],
    color='salmon'
)
axs[1].set_title('Emisiones de CO₂ por marca-año')

plt.tight_layout()
```

### Esta marca en que paises produce mas toneladas de ropa y CO2

```python
brand_produced_co2_country = (
    data[data['Brand']=='H&M']
    .groupby('Country')[['Monthly_Production_Tonnes', 'Carbon_Emissions_tCO2e']]
    .sum()
    .sort_values(by='Monthly_Production_Tonnes',ascending=False)
)
brand_produced_co2_country
```

### ¿Qué año tuvo el mayor impacto ambiental?

```python
year_impact_index_brand = data.groupby('Year')[['Carbon_Emissions_tCO2e','Water_Usage_Million_Litres','Landfill_Waste_Tonnes']].sum()
year_impact_index_brand
```

```python
fig, axs = plt.subplots(1,3, figsize=(14,4))

axs[0].bar(
    year_impact_index_brand.index, year_impact_index_brand['Carbon_Emissions_tCO2e'], color='firebrick'
)
axs[0].set_title('Emision de CO2')
axs[0].ticklabel_format(style='plain', axis='y') 

axs[1].bar(
    year_impact_index_brand.index, year_impact_index_brand['Water_Usage_Million_Litres'], color='steelblue'
)
axs[1].set_title('Emision de CO2')

axs[2].bar(
    year_impact_index_brand.index, year_impact_index_brand['Landfill_Waste_Tonnes'], color='darkslategray'
)
axs[2].set_title('Emision de CO2')
```
