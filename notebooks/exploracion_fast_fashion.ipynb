## Consultas basicas

### Qué país consume más ropa al año?

```python
country_prod_total = data.groupby('Country')['Monthly_Production_Tonnes'].sum().sort_values(ascending=False).head(5)
country_prod_total
```

```python
plt.plot(country_prod_total.index, country_prod_total.values, marker='o', linestyle='-', color='red')
plt.grid()

plt.title('Top 5 paises con mayor consumo de moda')
plt.xlabel('Country')
plt.ylabel('Monthly Production Tonnes')
plt.show()
```

### ¿Qué país genera más residuos textiles?

```python
country_res_total = data.groupby('Country')['Landfill_Waste_Tonnes'].sum().sort_values(ascending=False).head()
country_res_total
```

### Promedio de emisiones de CO₂ por país.

```python
avg_emision_co2 = data.groupby('Country')['Carbon_Emissions_tCO2e'].mean()
avg_emision_co2
```

### Top 5 paises con mayor producción de ropa

```python
country_more_production_clothes = data.groupby('Country')['Monthly_Production_Tonnes'].sum().sort_values(ascending=False).head(5)
country_more_production_clothes
```

## Analisis comparativo

### Comparar países desarrollados vs países en desarrollo (si hay categorización).

```python
#creando categorización de los paises desarrollados con los que estan en desarrollo
desarrollados = ['USA', 'UK', 'Germany', 'France', 'Japan', 'Canada']

data['Development'] = data['Country'].apply(lambda x: 'Desarrollado' if x in desarrollados else 'En desarrollo')
```

```python
data.info()
```

### ¿Los paises desarrollados producen mas ropa o los que estan en desarrollo?

```python
development_or_not_development = data.groupby('Development')['Monthly_Production_Tonnes'].sum().sort_values(ascending=False)
print(development_or_not_development)

plt.subplots(figsize=(8,6))
labels = ['Desarrollados', 'No desarrollados']
colors = ['lightblue', 'lightgreen']

plt.pie(development_or_not_development, labels=labels, colors=colors, autopct='%1.1f%%',startangle=140)

plt.title('Paises desarrollados VS No desarrollados')
plt.show()
```

### Paises que compran mas ropa al año (Segun su desarrollo)

```python
country_development = data[data['Development']== 'Desarrollado']
dev_grouped = country_development.groupby('Country')['Monthly_Production_Tonnes'].sum()
dev_grouped
```

```python
country_not_development = data[data['Development']== 'En desarrollo']
not_dev_grouped = country_not_development.groupby('Country')['Monthly_Production_Tonnes'].sum()
not_dev_grouped
```

```python
fig, axs = plt.subplots(1,2, figsize=(10, 4))
axs[0].bar(country_development['Country'], country_development['Monthly_Production_Tonnes'], color='blue')
axs[0].set_title('Producción mensual de ropa por país desarrollado')
axs[0].tick_params(axis='x', rotation=45)

axs[1].bar(country_not_development['Country'], country_not_development['Monthly_Production_Tonnes'], color='green')
axs[1].set_title('Producción mensual de ropa por país en desarrollo')
axs[1].tick_params(axis='x', rotation=45)
```

## Relación entre consumo de ropa y emisiones de CO₂.

### Marca que produce mas CO2 y ropa al año

```python
Brand_produced_co2 = data.groupby(['Year','Brand'])[['Monthly_Production_Tonnes', 'Carbon_Emissions_tCO2e']].sum().sort_values(by= 'Carbon_Emissions_tCO2e', ascending=False).head(10)
Brand_produced_co2
```

```python
import matplotlib.pyplot as plt

# Reiniciamos el índice para poder acceder a las columnas 'Year' y 'Brand' directamente
Brand_produced_co2 = Brand_produced_co2.reset_index()

# Tamaño general de las gráficas
plt.figure(figsize=(14, 6))

# --------- 1. Gráfico de producción de ropa ----------
plt.subplot(1, 2, 1)
plt.barh(
    Brand_produced_co2['Brand'] + ' (' + Brand_produced_co2['Year'].astype(str) + ')',
    Brand_produced_co2['Monthly_Production_Tonnes'],
    color='skyblue'
)
plt.xlabel('Toneladas de ropa producidas')
plt.title('Producción de ropa por marca-año')
plt.gca().invert_yaxis()  # Para que el mayor quede arriba

# --------- 2. Gráfico de emisiones de CO2 ----------
plt.subplot(1, 2, 2)
plt.barh(
    Brand_produced_co2['Brand'] + ' (' + Brand_produced_co2['Year'].astype(str) + ')',
    Brand_produced_co2['Carbon_Emissions_tCO2e'],
    color='salmon'
)
plt.xlabel('Emisiones de CO₂ (tCO2e)')
plt.title('Emisiones de CO₂ por marca-año')
plt.gca().invert_yaxis()

# Ajustar el diseño
plt.tight_layout()
plt.show()
```
