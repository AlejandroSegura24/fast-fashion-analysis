### Qué país consume más ropa al año?

```python
country_prod_total = data.groupby('Country')['Monthly_Production_Tonnes'].sum().sort_values(ascending=False).head(5)
country_prod_total
```

### ¿Qué país genera más residuos textiles?

```python
country_res_total = data.groupby('Country')['Landfill_Waste_Tonnes'].sum().sort_values(ascending=False).head()
country_res_total
```

### Promedio de emisiones de CO₂ por país.

```python
avg_emision_co2 = data.groupby('Country')['Carbon_Emissions_tCO2e'].mean()
avg_emision_co2
```

### Top 5 paises con mayor producción de ropa

```python
country_more_production_clothes = data.groupby('Country')['Monthly_Production_Tonnes'].sum().sort_values(ascending=False).head(5)
country_more_production_clothes
```

## 3. Analisis comparativo

### Comparar países desarrollados vs países en desarrollo (si hay categorización).

```python
#creando categorización de los paises desarrollados con los que estan en desarrollo
desarrollados = ['USA', 'UK', 'Germany', 'France', 'Japan', 'Canada']

data['Development'] = data['Country'].apply(lambda x: 'Desarrollado' if x in desarrollados else 'En desarrollo')
```

```python
data.head()
```

## Relación entre consumo de ropa y emisiones de CO₂.

### Marca que produce mas CO2 al año

```python
Brand_produced_co2 = data.groupby(['Year','Brand'])[['Monthly_Production_Tonnes', 'Carbon_Emissions_tCO2e']].sum().sort_values(by= 'Carbon_Emissions_tCO2e', ascending=False).head(10)
Brand_produced_co2
```

### Esta marca en que pais produce mas toneladas de ropa y CO2

```python
brand_produced_co2_country = data[data['Brand']=='H&M'].groupby('Country')[['Monthly_Production_Tonnes', 'Carbon_Emissions_tCO2e']].sum().sort_values(by='Monthly_Production_Tonnes',ascending=False)
brand_produced_co2_country
```

### ¿Qué año tuvo el mayor impacto ambiental por marca?

```python
year_impact_index_brand = data.groupby('Year')[['Carbon_Emissions_tCO2e','Water_Usage_Million_Litres','Landfill_Waste_Tonnes']].sum()
year_impact_index_brand
```
